```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import chi2_contingency
```

```python
data=pd.read_csv('/content/smoking.csv')
data.head()
```

```python
def chi_square_test(observed, expected):
  """Calculates the Chi-square statistic.

  Args:
    observed: A list or NumPy array of observed counts.
    expected: A list or NumPy array of expected counts.

  Returns:
    A float representing the Chi-square statistic.
  """

  # Calculate the squared deviations between the observed and expected counts.
  squared_deviations = (observed - expected)**2

  # Calculate the Chi-square statistic by dividing the squared deviations by the expected counts.
  chi_square_statistic = np.sum(squared_deviations / expected)
```

```python
def chi_square_test(observed):
    # Calculate the chi-square statistic and p-value using scipy's chi2_contingency
    chi2, p, _, _ = stats.chi2_contingency(observed)

    # Calculate degrees of freedom
    dof = (observed.shape[0] - 1) * (observed.shape[1] - 1)

    # Print the results
    print("Chi-Square Statistic:", chi2)
    print("Degrees of Freedom:", dof)
    print("P-value:", p)

    # Determine if the result is significant
    alpha = 0.05  # Set your desired significance level
    if p < alpha:
        print("The result is statistically significant (reject the null hypothesis).")
    else:
        print("The result is not statistically significant (fail to reject the null hypothesis).")
```

```python
# Assuming your dataset has two categorical columns: 'gender' and 'smoking'
observed = pd.crosstab(data['gender'], data['smoking'])
```

```python
chi_square_test(observed)
```

```python
_, p_standard, _, _ = stats.chi2_contingency(observed)
print("P-value (Standard Function):", p_standard)
```

```python
# Create a heatmap to visualize the observed frequencies
plt.figure(figsize=(8, 6))
sns.heatmap(observed, annot=True, cmap='coolwarm', fmt='d', cbar=False)
plt.xlabel("Outcome")
plt.ylabel("Smoker")
plt.title("Chi-Square Test - Observed Frequencies")
plt.show()
```
