```python
import numpy as np
import pandas as pd
```

```python
# 1.Read the three csv files which contains the score of same students in term1 for each Subject
df1=pd.read_csv('/content/DSScoreTerm1 - DSScoreTerm1.csv')
df2=pd.read_csv('/content/MathScoreTerm1 - MathScoreTerm1.csv')
df3=pd.read_csv('/content/PhysicsScoreTerm1 - PhysicsScoreTerm1.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
# 2.Remove the name and ethnicity column (to ensure confidentiality)
df1=df1.drop(['Name','Ethinicity'],axis=1)
```

```python
df2=df2.drop(['Name','Ethinicity'],axis=1)
df3=df3.drop(['Name','Ethinicity'],axis=1)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
# 3.Fill missing score data with zero
print(df1.isna().sum())
print(df2.isna().sum())
print(df3.isna().sum())
```

```python
df1=df1.fillna(0)
df2=df2.fillna(0)
df3=df3.fillna(0)
```

```python
print(df1.isna().sum())
print(df2.isna().sum())
print(df3.isna().sum())
```

```python
#4. Merge two files
# using merge function by setting how='inner'
out1= pd.merge(df1, df2,
                   on=['ID','Age','Sex'],
                   how='inner')
```

```python
out2=pd.merge(df3,out1,on=['ID','Age','Sex'],how='inner')
```

```python
out2.head()
```

```python
# 4 Change Sex(M/F) Column to 1/2 for further analysis
out2=out2.replace('M',1)
out2=out2.replace('F',2)
```

```python
out2.head()
```

```python
# 6. Store the data in new file â€“ScoreFinal.csv
out2.to_csv('ScoreFinal.csv')
```

```python
# 7.Convert ethnicity to numerical value
from sklearn import preprocessing
encoder=preprocessing.LabelEncoder()

# Encode labels in column 'species'.
df3=pd.read_csv('/content/PhysicsScoreTerm1 - PhysicsScoreTerm1.csv')
df3['Ethinicity']= encoder.fit_transform(df3['Ethinicity'])
df3.head()
```

```python
#8.Fill the missing score for a student to the average of the class
df3.isna().sum()
```

```python
df3['Score']=df3['Score'].fillna(df3['Score'].mean())
```

```python
df3.isna().sum()
```
