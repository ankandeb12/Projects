Performing Linear Regression and Multiple Linear Regression on 2 different datasets, cars.csv and kc_house_data.csv available on kaggle

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Data Set 1
x1 = np.array([0, 2, 3, 5, 6, 7, 8, 8, 9, 10])
y1 = np.array([2, 4, 5, 6, 7.5, 8.5, 9, 10, 11, 12])

# (a) Plot the data
plt.scatter(x1, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Set 1')
plt.show()

# (b) Linear Regression
x1 = x1.reshape(-1, 1)
model1 = LinearRegression().fit(x1, y1)

# (c) Predict a value
x_new = np.array([[4]])  # Example value for prediction
y_pred = model1.predict(x_new)
print(f'(c) Prediction for x=4: {y_pred[0]}')

# (d) Show the linear regression line
plt.scatter(x1, y1, label='Data')
plt.plot(x1, model1.predict(x1), color='red', label='Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Set 1 with Linear Regression Line')
plt.legend()
plt.show()
```

```python
# Data Set 2
x2 = np.array([72, 50, 81, 74, 94, 86, 59, 83, 65, 33, 88, 81])
y2 = np.array([84, 63, 77, 78, 90, 75, 49, 79, 77, 52, 74, 90])

# (a) Plot the data
plt.scatter(x2, y2)
plt.xlabel('Midterm Exam Grade')
plt.ylabel('Final Exam Grade')
plt.title('Data Set 2')
plt.show()

# (b) Method of Least Squares
def least_squares(x, y):
    n = len(x)
    mean_x, mean_y = np.mean(x), np.mean(y)
    slope = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x) ** 2)
    intercept = mean_y - slope * mean_x
    return slope, intercept

slope, intercept = least_squares(x2, y2)
print(f'(b) Equation: y = {slope:.2f}x + {intercept:.2f}')

# (c) Plot with the linear equation line
plt.scatter(x2, y2, label='Data')
plt.plot(x2, slope * x2 + intercept, color='red', label='Linear Regression')
plt.xlabel('Midterm Exam Grade')
plt.ylabel('Final Exam Grade')
plt.title('Data Set 2 with Linear Regression Line')
plt.legend()
plt.show()

# (d) Predict for x=86
x_new2 = 86
y_pred2 = slope * x_new2 + intercept
print(f'(d) Prediction for Midterm Grade = 86: {y_pred2}')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Part 3: Multiple Linear Regression on cars.csv Dataset

# (a) Analyze each column of the dataset using appropriate visualization techniques.
cars_df = pd.read_csv('cars.csv')
sns.pairplot(cars_df)
plt.show()

# (b) Consider only weight and volume columns as multiple variables to predict the CO2 emission.
X_cars = cars_df[['Weight', 'Volume']]
y_cars = cars_df['CO2']

# (c) Predict the CO2 emission for a car with weight=2300kg and volume=1300ccm.
car_to_predict = [[2300, 1300]]
model_cars = LinearRegression().fit(X_cars, y_cars)
co2_prediction = model_cars.predict(car_to_predict)
print(f'(c) Predicted CO2 Emission: {co2_prediction[0]}')

# (d) Perform predictions on a few more values.
additional_cars = [[2400, 1400], [2000, 1200]]
co2_predictions_additional = model_cars.predict(additional_cars)
print(f'(d) Predicted CO2 Emission for Additional Values: {co2_predictions_additional}')

# (e) Show the coefficients obtained and conclude.
coefficients_cars = model_cars.coef_
intercept_cars = model_cars.intercept_
print(f'(e) Coefficients: {coefficients_cars}, Intercept: {intercept_cars}')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Part 4: Linear Regression on kc_house_data.csv Dataset

# (a) Load the dataset, display it, visualize various columns, and explain the dataset composition.
house_df = pd.read_csv('kc_house_data.csv')
print(house_df.head())
sns.pairplot(house_df[['price', 'bedrooms', 'bathrooms', 'sqft_living']])
plt.show()

# (b) Describe the dataset.
description_house = house_df.describe()
print(f'(b) Dataset Description:\n{description_house}')

# (c) Houses with how many bedrooms are most sold?
most_sold_bedrooms = house_df['bedrooms'].value_counts().idxmax()
print(f'(c) Houses with {most_sold_bedrooms} bedrooms are most sold.')

# (d) Visualizing the location of houses based on latitude and longitude.
plt.scatter(house_df['long'], house_df['lat'], alpha=0.3, edgecolors='w', s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Location of Houses')
plt.show()

# (e) Find correlated features/columns.
correlation_matrix_house = house_df.corr()
correlated_features_house = correlation_matrix_house['price'].sort_values(ascending=False).index[1:]
print(f'(e) Correlated Features with Price:\n{correlated_features_house}')

# (f) Find null values and fill with mean for all columns.
house_df.fillna(house_df.mean(), inplace=True)

# (g) Find dependent and independent data (place in X and y).
X_house = house_df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']]
y_house = house_df['price']

# (h) Split train and test data.
X_train_house, X_test_house, y_train_house, y_test_house = train_test_split(X_house, y_house, test_size=0.2, random_state=42)

# (i) Train the model and test it. Find the accuracy.
model_house = LinearRegression()
model_house.fit(X_train_house, y_train_house)
accuracy_house = model_house.score(X_test_house, y_test_house)
print(f'(i) Model Accuracy: {accuracy_house}')

# (j) Test the model using some arbitrary input.
arbitrary_input_house = np.array([[3, 2, 2000, 5000, 2]])
price_pred_house = model_house.predict(arbitrary_input_house)
print(f'(j) Predicted Price for Arbitrary Input: {price_pred_house[0]}')
```
